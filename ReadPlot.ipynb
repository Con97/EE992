```python
def filter_lst(lst):
    filter_lst = [x for x in lst if ".csv" in x]
    return filter_lst

def filter_model_names(lst, model):
    filter_lst = [x for x in lst if model in x]
    return filter_lst

# def get_model_perf(filtered_files):
#     max_acc = 0
#     max_file = ""
# for file in filtered_files:
#     acc = max(pd.read_csv(folder+file)["val_accuracy"])
#     if acc > max_acc:
#         max_file = file
#         max_acc = acc
    

def lst_to_df(lst, folder):
    """"Takes in list of model .csv files and
        retruns a df of the model hyperparemters, 
        max acc and min loss
        """
    df = pd.DataFrame(columns=['Model', 'Dropout', 'Learning Rate', 'Training', 'Accuarcy', 'Loss'])
    
    for file in lst:

        acc = max(pd.read_csv(folder+file)["val_accuracy"])
        loss = min(pd.read_csv(folder+file)["val_loss"])
        
        file = file[:-4]
        file_split = file.split("_")
        
        model = file_split[0]
        dropout = float(file_split[1])
        lr = float(file_split[2])
        training = file_split[3]

        data = [model, dropout, lr, training, acc, loss]

        df = pd.concat([df, pd.DataFrame([data], columns=df.columns)], ignore_index=True)

    return df
```
